```python
import numpy as np
import matplotlib.pyplot as plt

def rk2(f, y0, x0, h, n):
  """
  Runge-Kutta 2nd order method

  Args:
    f: The function defining the differential equation.
    y0: The initial value of y.
    x0: The initial value of x.
    h: The step size.
    n: The number of steps.

  Returns:
    A tuple (x, y) where x is an array of x values and y is an array of corresponding y values.
  """

  x = np.zeros(n+1)
  y = np.zeros(n+1)
  x[0] = x0
  y[0] = y0

  for i in range(n):
    k1 = h * f(x[i], y[i])
    k2 = h * f(x[i] + h/2, y[i] + k1/2)
    y[i+1] = y[i] + k2
    x[i+1] = x[i] + h

  return x, y

# Define the differential equation
def dydx(x, y):
  return np.log(x*y)

# Initial condition
y0 = 1
x0 = 1
h = 0.1
n = 100

# Solve the differential equation using RK2
x, y = rk2(dydx, y0, x0, h, n)

# Plot the solution
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution to dy/dx = log(xy) using RK2')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(np.abs(x-1))

def df_dx(x, y):
    return np.cos(np.abs(x-1)) * np.sign(x-1)

def rk4(f, y0, x0, h, n):
    """
    Runge-Kutta 4th order method
    """
    x = np.zeros(n + 1)
    y = np.zeros(n + 1)
    x[0] = x0
    y[0] = y0

    for i in range(n):
        k1 = h * f(x[i], y[i])
        k2 = h * f(x[i] + h / 2, y[i] + k1 / 2)
        k3 = h * f(x[i] + h / 2, y[i] + k2 / 2)
        k4 = h * f(x[i] + h, y[i] + k3)
        y[i + 1] = y[i] + (k1 + 2 * k2 + 2 * k3 + k4) / 6
        x[i + 1] = x[i] + h

    return x, y

# Parameters
x0 = 0
y0 = np.sin(1)
h = 0.01
n = 1000

# Numerical solution using RK4
x_rk4, y_rk4 = rk4(df_dx, y0, x0, h, n)

# Analytical solution
x_analytical = np.linspace(0, 4, 100)
y_analytical = f(x_analytical)

# Plot both solutions
plt.plot(x_analytical, y_analytical, label='Analytical')
plt.plot(x_rk4, y_rk4, label='RK4')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
